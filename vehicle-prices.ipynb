```python
# Import modules
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

#  Import and read Australian Vehicle Prices.csv

Vehicleprices_df = pd.read_csv("Australian Vehicle Prices.csv")
Vehicleprices_df.head()
```

```python

Vehicleprices_df.info()
```

```python
# Drop Columns 
columns_to_drop = ['Seats','Doors','ColourExtInt','DriveType','Title']
Vehicleprices_df = Vehicleprices_df.drop(columns=columns_to_drop)
Vehicleprices_df = Vehicleprices_df.loc[Vehicleprices_df['Price'] != 'POA']
Vehicleprices_df

```

```python
Vehicleprices_df = Vehicleprices_df.dropna()
```

## Separate the features X from the target y

```python
 # Seperate the features, X,  from the target variable, y
y = Vehicleprices_df['Price']
X = Vehicleprices_df.drop(columns='Price')
```

```python
# Preview the features data
X.head()
```

```python
 # Preview the first five entries for the target variable
y[:5]
```

## Encode the categorical variables from the features data using get_dummies.

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
```

```python
 # Review the features data
X.head()
```

## Separate the data into training and testing subsets.

```python
# Split the dataset using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

## Scale the data using StandardScaler

```python
# Instantiate a StandardScaler instance
scaler = StandardScaler()

# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)


import numpy as np

# En lugar de float64, puedes utilizar float32
array = np.zeros((11964, 16331), dtype=np.float32)
```

## Instantiate an K Nearest Neighbour Classifier instance.

```python
# Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNeighborsClassifier model with n_neighbors = 3 
knn = KNeighborsClassifier(n_neighbors=3)
```

## Fit the model using the training data.

```python
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
```

##  Make predictions using the testing data.

```python
 # Create predictions using the testing data
y_pred = knn.predict(X_test_scaled)
```

## Generate the classification report for the test data.

```python
# Print the classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred))
```
