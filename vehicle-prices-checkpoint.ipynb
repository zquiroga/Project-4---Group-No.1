```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read Australian Vehicle Prices.csv
import pandas as pd 
Vehicleprices_df = pd.read_csv("Australian Vehicle Prices.csv")
Vehicleprices_df.head()
```

```python

Vehicleprices_df.info()
```

```python
# Drop Columns 
columns_to_drop = ['Seats','Doors','ColourExtInt','DriveType','Title']
Vehicleprices_df = Vehicleprices_df.drop(columns=columns_to_drop)
Vehicleprices_df

```

```python
# Convert the 'Year' column to datetime type for easier comparison
Vehicleprices_df['Year'] = pd.to_datetime(Vehicleprices_df['Year'], format='%Y')

# Filter the rows with a year greater than 2020
Vehicleprices_df = Vehicleprices_df[Vehicleprices_df['Year'].dt.year > 2020]
Vehicleprices_df
```

```python
# Determine the number of unique values in each column.
Vehicleprices =Vehicleprices_df.dtypes[Vehicleprices_df.dtypes == "object"].index.tolist()
Vehicleprices_df[Vehicleprices].nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
Titles = Vehicleprices_df['Price'].value_counts().reset_index()
Titles
```

```python
# Look at APPLICATION_TYPE value counts for binning
Brands = Vehicleprices_df['Brand'].value_counts().reset_index()
Brands 
```

```python


# Choose a cutoff value and create a list of Brand types to be replaced
Brand_types_to_replace = []

Brand_types_to_replace = Brands[Brands['Brand'] <=27]['index']

# Replace in dataframe
for Brands in Brand_types_to_replace:
    Vehicleprices_df['Brand'] = Vehicleprices_df['Brand'].replace(Brands,"Other")

# Check to make sure binning was successful
Vehicleprices_df['Brand'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#Kilometres = Vehicleprices_df['Kilometres'].value_counts().reset_index()
#Kilometres
```

```python
# Determine the columns to be encoded
columns_to_encode = Vehicleprices_df.dtypes[Vehicleprices_df.dtypes == "object"].index.tolist()

# Encode categorical data
numerical_df = pd.get_dummies(Vehicleprices_df, columns=columns_to_encode)

# Sum the encoded 'Price' columns to get a single column
numerical_df['Price'] = numerical_df.filter(like='Price_').sum(axis=1)

# Drop the original encoded 'Price' columns
numerical_df = numerical_df.drop(numerical_df.filter(like='Price_').columns, axis=1)
numerical_df
```

```python
# Split our preprocessed data into our features and target arrays
y = numerical_df['Price'].values
X = numerical_df.drop(['Price'],1).values

# Normalize the features using StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=80)

```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 20

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="LeakyReLU"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="LeakyReLU"))

# Output layer with linear activation for regression
nn.add(tf.keras.layers.Dense(units=1, activation="linear"))


# Check the structure of the model
nn.summary()

```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```
